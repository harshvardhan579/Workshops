```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style; style.use('ggplot')
```

```python
# load data
data = pd.read_csv('linear_regression_data.csv')

# setup variables
X = data.X
y = data.y
m = y.size
```

```python
# visualize data
plt.scatter(X, y, color='blue')
plt.show()
```

```python
# add column of ones to X to account for y-intercept
X = np.hstack((np.ones((m, 1)), 
               X.values.reshape(m, 1)))

# make y 2-Dimensional to match X
y = y.values.reshape(m, 1)

# initialize parameters slope and y-intercept to 0
theta = np.zeros((2, 1))
```

```python
# function to make predictions
def predict():
    return np.matmul(X, theta)
```

```python
# function to compute cost
def cost():
    return (1 / (2 * m)) * np.sum((predict() - y) ** 2)
```

```python
# set gradient descent parameters
num_iterations = 1500
learning_rate = 0.01
```

```python
# record costs to plot later
cost_ls = []

# perform gradient descent
for i in range(num_iterations):
    gradient = (1 / m) * np.matmul(X.T, (predict() - y))
    theta -= learning_rate * gradient
    cost_ls.append(cost())
```

```python
# plot cost
plt.plot(cost_ls, color='red')
plt.show()
```

```python
# create endpoints for plotting
X_pt = [min(X[:, 1]), max(X[:, 1])]
y_pt = [min(predict()), max(predict())]

# plot line of best fit, learned through gradient descent
plt.scatter(X[:, 1], y, color='blue')
plt.plot(X_pt, y_pt, color='green')
plt.show()
```
